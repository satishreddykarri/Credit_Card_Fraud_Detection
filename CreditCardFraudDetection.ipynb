```python
#Credit Card Fraud Detection
#Importing the dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#balancing the dataset
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the vlaues for both transaction
credit_data.groupby('Class').mean()
```

```python
#undersampling
#building a sample dataset containing similar distribution of normal transactions and fradulent transactions
legit_sample = legit.sample(n=492)
```

```python
#concatinating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
acur = accuracy_score(y_test, y_pred)
```

```python
print(acur)
```
